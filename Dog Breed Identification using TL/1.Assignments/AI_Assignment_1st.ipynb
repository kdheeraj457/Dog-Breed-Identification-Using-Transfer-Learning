2. Write a function to Calculate Area and Perimeter of Rectangle

```python
def Rectangle_Properties(Length, Width):
    area = Length * Width
    perimeter = 2 * (Length + Width)
    return area, perimeter

Length = float(input('Length value of a Rectangle: '))
Width = float(input('Width value of a Rectangle: '))

Area, Perimeter = Rectangle_Properties(Length, Width)

print("Area:", Area)
print("Perimeter:", Perimeter)
```

3. Write a python Program to find out given No is Prime or Not.

```python
def is_prime(number):
    if number <= 10:
        return False
    elif number <= 3:
        return True
    elif number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

# Test the function
num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

5. Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price
First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)

```python
def calculate_electricity_bill(units):
    if units <= 100:
        return 0
    elif units <= 200:
        bill = (units - 100) * 5
        return bill
    else:
        bill = (100 * 5) + ((units - 200) * 10)
        return bill

# Prompt the user to input the number of units
units = int(input("Enter the number of units consumed: "))

# Calculate the electricity bill
total_bill = calculate_electricity_bill(units)

# Print the total bill amount
print("Total bill amount is:", total_bill)
```

6. Write a program to accept numbers from 1-7 and display name of day like- 1 for sunday, 2 for Monday and so on.

```python
day_names = {
  1: "Sunday",
  2: "Monday",
  3: "Tuesday",
  4: "Wednesday",
  5: "Thursday",
  6: "Friday",
  7: "Saturday"
}

# Get the day number (assuming it's between 1 and 7)
while True:
  try:
    day_number = int(input("Enter a number between 1 and 7: "))
    if 1 <= day_number <= 7:
      break
    else:
      print("Invalid input. Please enter a number between 1 and 7.")
  except ValueError:
    print("Invalid input. Please enter a number.")

# Print the day name
print(f"The day is {day_names[day_number]}.")
```

7. Write a python program to check given word is Palindrome or not

```python
def is_palindrome(word):
    # Convert the word to lowercase
    word = word.lower()
    # Reverse the word
    reversed_word = word[::-1]
    # Check word is equal to its reverse
    if word == reversed_word:
        return True
    else:
        return False

word = input("Write a word: ")

if is_palindrome(word):
    print(word, "is a palindrome.")
else:
    print(word, "is not a palindrome.")
```

10. Perform Bivariate Analysis

```python
# Bivariate Analysis - Analyze relationship between two columns
print("\nBivariate Analysis:")
numerical_cols = data.select_dtypes(include=[np.number])  # Select numerical columns
for col1 in numerical_cols.columns:
  for col2 in numerical_cols.columns:
    if col1 != col2:
      print(f"Relationship between {col1} and {col2}")
      print(data[[col1, col2]].corr())
```
